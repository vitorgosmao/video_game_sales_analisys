# Análise dos jogos mais vendidos por plataforma ou gênero

### Top 10 jogos mais vendidos por plataforma

```python
top_platforms = df.groupby('Platform')['Global_Sales'].sum().nlargest(10)
top_games_by_platform = df.loc[df['Platform'].isin(top_platforms.index)]
top_games_by_platform.groupby(['Platform', 'Name'])['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

### Top 10 jogos mais vendidos por gênero

```python
top_genres = df.groupby('Genre')['Global_Sales'].sum().nlargest(10)
top_games_by_genre = df.loc[df['Genre'].isin(top_genres.index)]
top_games_by_genre.groupby(['Genre', 'Name'])['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

# Comparação das vendas globais por gênero ao longo dos anos

### Total de vendas globais por ano e por gênero

```python
sales_by_genre_year = df.groupby(['Year_of_Release', 'Genre'])['Global_Sales'].sum().reset_index()
```

```python
sns.lineplot(x='Year_of_Release', y='Global_Sales', hue='Genre', data=sales_by_genre_year)
plt.title('Vendas globais por gênero ao longo dos anos')
plt.show()

## melhorar essa visualização
```

# Identificação das editoras com mais jogos lançados e as suas respectivas médias de vendas

### Editoras com mais jogos lançados

```python
top_publishers = df['Publisher'].value_counts().nlargest(10).index.tolist()
top_publishers
```

### Média de vendas globais por editora

```python
publisher_sales = df.loc[df['Publisher'].isin(top_publishers)].groupby('Publisher')['Global_Sales'].mean().sort_values(ascending=False)
publisher_sales.plot(kind='bar')
plt.title('Média de vendas globais por editora')
plt.show()
```

# Análise das vendas por região para identificar as preferências dos jogadores em diferentes áreas geográficas

### Vendas por região

```python
sales_by_region = df.groupby(['Genre'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
```

### Heatmap das vendas por região

```python
sns.heatmap(sales_by_region, cmap='Blues')
plt.title('Vendas por região')
plt.show()
```
